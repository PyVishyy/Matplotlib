```python
#This is example for matplotlib normal graph.
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 4, 6])
y = np.array([5, 8, 3])

plt.plot(x, y)
plt.show()
```

```python
#This is matplotlib graph basic example.
import matplotlib.pyplot as plt
import numpy as np

plt.title("Basic matplotlib")

x = np.array([7, 8 ,1 ,5])
y = np.array([4, 3, 7, 9])

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.plot(x, y)
plt.show()
```
